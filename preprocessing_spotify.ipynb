```python
# Mengganti nama kolom menggunakan dictionary yang telah didefinisikan sebelumnya
rename_dict = {
    "Sebutkan judul musik dan artis yang sering didengarkan ketika mengerjakan skripsi (Top 1)": "musik1",
    "Sebutkan judul musik dan artis yang sering didengarkan ketika mengerjakan skripsi (Top 2)": "musik2",
    "Sebutkan judul musik dan artis yang sering didengarkan ketika mengerjakan skripsi (Top 3)": "musik3",
    "Sebutkan judul musik dan artis yang sering didengarkan ketika mengerjakan skripsi (Top 4)": "musik4",
    "Sebutkan judul musik dan artis yang sering didengarkan ketika mengerjakan skripsi (Top 5)": "musik5",
}
data.rename(columns=rename_dict, inplace=True)
```

```python
# Hapus baris berdasarkan kondisi tertentu dan hapus kolom yang tidak diperlukan
data = data[data['Apakah anda sedang menyusun skripsi di tahun 2023?'] == 'Ya']
data.drop('Apakah anda sedang menyusun skripsi di tahun 2023?', axis=1, inplace=True)
```

```python
# Hapus baris berdasarkan kondisi tertentu dan hapus kolom yang tidak diperlukan
data = data[data['Apakah anda merasa musik membantu anda dalam meningkatkan konsentrasi saat menulis skripsi?'] == 'Ya']
data.drop('Apakah anda merasa musik membantu anda dalam meningkatkan konsentrasi saat menulis skripsi?', axis=1, inplace=True)
```

```python
# Perform casefolding to convert all responses to lowercase
data = data.applymap(lambda s: s.lower() if type(s) == str else s)
```

```python
# Menghitung jumlah responden untuk setiap kategori
kategori_musik = data['Seberapa sering anda mendengarkan musik saat menyusun skripsi?'].value_counts()

# Mengurutkan kategori sesuai dengan permintaan
kategori_urut = kategori_musik.reindex([
    "jarang (1-3 jam/hari)",
    "kadang-kadang (4-6 jam/hari)",
    "sering ( > 6 jam/hari)" 
])

# Pastikan kategori_urut tidak mengandung NaN dengan mengisi nilai NaN dengan 0
kategori_urut = kategori_urut.fillna(0)

# Membuat bar chart
plt.figure(figsize=(10, 6))
# Mengganti warna bar di sini
bars = plt.bar(kategori_urut.index, kategori_urut.values, color=['#CF0A0A', '#FFDE00', '#5FD068'])

# Menambahkan judul dan label
#plt.title('Frekuensi Mendengarkan Musik Saat Menyusun Skripsi', fontsize=15)
plt.xlabel('Frekuensi Mendengarkan Musik', fontsize=12)
plt.ylabel('Jumlah Responden', fontsize=12)

# Memastikan skala sumbu Y sesuai dengan nilai maksimum dari jumlah responden
plt.ylim(0, kategori_urut.values.max() + 10)  # Tambahkan sedikit ruang di atas bar tertinggi

# Menampilkan nilai pada setiap bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom', ha='center', fontsize=10)

# Menampilkan grafik
plt.show()
```

```python
# MENGETAHUI URUTAN MUSIK YANG PALING BANYAK DIDENGAR RESPONDEN
# First, we extract all the music-related columns which are 'musik1' through 'musik5'
music_columns = ['musik1', 'musik2', 'musik3', 'musik4', 'musik5']

# Menggabungkan semua respons musik ke dalam satu array dan menghapus nilai NaN
music_responses = data[music_columns].stack().dropna()

# Create a series from the music responses to utilize value_counts
music_series = pd.Series(music_responses)

# Count the frequency of each song
music_frequency = music_series.value_counts()

# Display the sorted music frequency
music_frequency.head(11)
#music_frequency
```

```python
music_frequency.count()
```

```python
# Menggabungkan semua kolom 'musik' ke dalam satu Series
all_songs = pd.concat([data['musik1'], data['musik2'], data['musik3'], data['musik4'], data['musik5']])

# Menghitung frekuensi setiap lagu
song_frequency = all_songs.value_counts()

# Menampilkan frekuensi lagu
print(song_frequency)
```

```python
# Menyimpan hasil ke file Excel
# Ganti 'song_frequency.xlsx' dengan nama file yang Anda inginkan
output_file = 'song_frequency.xlsx'
song_frequency.to_excel(output_file, index=True, header=["Frequency"])
```

```python
# Membagi lagu berdasarkan frekuensi dan menyimpan ke dalam file Excel terpisah
frekuensi_1 = song_frequency[song_frequency == 1]
frekuensi_2 = song_frequency[song_frequency == 2]
frekuensi_3 = song_frequency[song_frequency == 3]
frekuensi_lebih_dari_3 = song_frequency[song_frequency > 3]

# Menyimpan hasil ke dalam file Excel
frekuensi_1.to_excel('frekuensi_1.xlsx', index=True, header=["Frequency"])
frekuensi_2.to_excel('frekuensi_2.xlsx', index=True, header=["Frequency"])
frekuensi_3.to_excel('frekuensi_3.xlsx', index=True, header=["Frequency"])
frekuensi_lebih_dari_3.to_excel('frekuensi_lebih_dari_3.xlsx', index=True, header=["Frequency"])

print("Pembagian lagu berdasarkan frekuensi telah disimpan dalam file Excel terpisah.")
```

```python
# Pastikan top_tracks sudah tidak memiliki duplikasi
top_tracks = list(dict.fromkeys(top_tracks))

audio_features_list = []
track_info_list = []  # List untuk menyimpan informasi lagu

for track in top_tracks:
    try:
        # Search for the track on Spotify, adding artist name to improve accuracy
        result = sp.search(q='artist:' + track.split('_')[0] + ' track:' + track.split('_')[1], type="track", limit=1)
        
        if result['tracks']['items']:
            track_item = result['tracks']['items'][0]
            track_id = track_item['id']

            # Fetch the audio features for the track
            audio_features = sp.audio_features([track_id])[0]

            if audio_features:
                audio_features_list.append(audio_features)
                track_info_list.append(track)  # Menyimpan informasi lagu
        else:
            print(f"No results for '{track}'")
    except Exception as e:
        print(f"An error occurred: {e}")

# Convert the list of dictionaries to a DataFrame
audio_features_df = pd.DataFrame(audio_features_list)

# Menambahkan kolom informasi lagu
audio_features_df['track_info'] = top_tracks

# Display the DataFrame
print(audio_features_df)
```
